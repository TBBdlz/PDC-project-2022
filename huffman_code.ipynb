```python
from sage.coding.source_coding.huffman import Huffman
from collections import OrderedDict

# Define the probability distribution
probs = [0.00817517, 0.0244164, 0.0427674, 0.060525, 0.0760872, 0.0887247, 0.0974304,
         0.101874, 0.101874, 0.0974304, 0.0887247, 0.0760872, 0.060525,
         0.0427674, 0.0244164, 0.00817517]

P = OrderedDict(('x{}'.format(i+1), p) for i, p in enumerate(probs))

# Print the probability distribution
print("Probability of each symbol in distribution P:")
for symbol, prob in P.items():
    print(f"{symbol}: {prob}")

# Compute the Huffman code for the given distribution
h = Huffman(P)

# Print the Huffman code
print("\nHuffman code for distribution P:")
for symbol, code in sorted(h.encoding_table().items()):
    print(f"{symbol}: {code}")
```

```python
print("Compute average code length")
for symbol, prob in P.items():
    print(f"{symbol}\t{prob}\t{h.encoding_table()[symbol]}")

R_bar = 0
for symbol, prob in P.items():
    R_bar += prob * len(h.encoding_table()[symbol])
print(f"R avg = {R_bar}")
```

```python
from sage.coding.source_coding.huffman import Huffman
from collections import OrderedDict

probs = [0.00817517, 0.0244164, 0.0427674, 0.060525, 0.0760872, 0.0887247, 0.0974304,
         0.101874, 0.101874, 0.0974304, 0.0887247, 0.0760872, 0.060525,
         0.0427674, 0.0244164, 0.00817517]

# Create a dictionary mapping characters to probabilities
P = OrderedDict(('x{}'.format(i+1), p) for i, p in enumerate(probs))

# Print the probability of each character
print("Probability of each character:")
for k, v in P.items():
    print(k, ": ", v)

# Create a Huffman code for the given frequency distribution
huff = Huffman(P)

# Get the Huffman tree and plot it
T = huff.tree()
plot = T.plot(figsize=[15, 15], layout='tree', tree_orientation='right', vertex_size=400)
plot.show()
```
