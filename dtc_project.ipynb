```julia
using Plots;
using LaTeXStrings;
using LinearAlgebra;
using QuadGK;
```

```julia
id_1 = 6_330_401_726
id_2 = 6_330_401_742
x0 = (id_1 + id_2) % 100
println("x₀ = $x0")
```

```julia
function xn(n::Int64)
    if (n == 0) # basecase return x0
        return x0
    elseif (n == 1) # basecase return x1 = 7
        return 7
    elseif (n >= 2)
        return (3 * xn(n-1) + 2 * xn(n-2)) % 19 + 2
    else
        error("Parameter n must be non-negative")
    end 
end
```

```julia
# test function xn
for i in 0:10
    println("x$i = $(xn(i))")
end
```

```julia
x2 = xn(2)

# define power-spectral function
function Sx(f)
    if (f > -120 && f < 120)
        return 2 + ((5^x2 % 23) / 25)
    else
        return 0
    end
end
```

```julia
# plot to check shape of Sx(f)
plotly() # use plotly as backend
f = -160:0.1:160
y = Sx.(f)

plot(f, y, label="Sx(f)", figsize=(8, 2), ylim=(0, 3))
xlabel!("f")
ylabel!("Sx(f)")
title!("Plot of Sx(f)")
```
